# Check Validation State

### 학습 과정을 파악하기 위하여 Validation set의 log 파일을 읽고, PSNR의 추세를 확인한다.

### - Class 1 (easy) average PSNR :  37.72226148769038
### - Class 2 (middle) average PSNR :  26.869776719975032
### - Class 3 (hard) average PSNR :  16.968756576079898

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_val_panr(path):
    f = open(path,'r')
    lines = f.readlines()

    psnr_log = []
    title = lines[0].split("name: ")[1]

    for line in lines:
        if "Validation" in line:
            psnr = float(line.split("PSNR: ")[1])
            psnr_log.append(psnr)
            
    x_values = np.arange(0,len(psnr_log))
    y_values = psnr_log
    max_panr = np.max(psnr_log)
    plt.plot(x_values, y_values)
    plt.title("PSNR (Validation) - %s"%title)
    plt.xlabel("iteration (5000 iter)")
    plt.ylabel("PSNR")
    plt.show()
    print("Best PSNR : %.4f"%max_panr)
    
def visualize_val_panr_epoch(path):
    f = open(path,'r')
    lines = f.readlines()

    psnr_log = []
    title = lines[0].split("name: ")[1]

    for line in lines:
        if "Validation" in line:
            psnr = float(line.split("PSNR: ")[1])
            psnr_log.append(psnr)
            
    x_values = np.arange(0,len(psnr_log))
    y_values = psnr_log
    max_panr = np.max(psnr_log)
    plt.plot(x_values, y_values)
    plt.title("PSNR (Validation) - %s"%title)
    plt.xlabel("epoch")
    plt.ylabel("PSNR")
    plt.show()
    print("Best PSNR : %.4f"%max_panr)
```

## SR Branch 1 (FSRCNN - 16)

```python
# Validation check 단위를 epoch 기준으로 변경 필요
path = "./experiments/FSRCNN_branch1/train_FSRCNN_branch1_210707-174922.log"
visualize_val_panr(path)
```
